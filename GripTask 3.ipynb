In our dataset the country columns containing only one value so we will drop Country Column

```python
data1=data.drop(['Country'],axis=1)
```

```python
data1.head()
```

```python
data1.hist(figsize=(12,12))
```

```python
sns.countplot('Ship Mode',data=data1,hue="Segment",palette="plasma")
```

```python
sns.countplot('Ship Mode',data=data1,hue='Region',palette="tab20c")
```

```python
fig, axes = plt.subplots(1,2, figsize = (15, 5))
data1.groupby("Ship Mode")["Sales"].sum().sort_values(ascending=False)[:10].plot(kind="bar",color="maroon",ax=axes[0]).set_title('Shipment Wise Sell')
axes[0].set_ylabel("Sales")
data1.groupby("Ship Mode")["Profit"].sum().sort_values(ascending=False)[:10].plot(kind="bar",color="lawngreen",ax=axes[1]).set_title('Shipment Wise profit')
axes[1].set_ylabel("Profit")
```

```python
fig, axes = plt.subplots(1,2, figsize = (15, 5))
data1.groupby("Segment")["Sales"].sum().sort_values(ascending=False)[:10].plot(kind="bar",color="navy",ax=axes[0]).set_title('Segment Wise Sell')
axes[0].set_ylabel("Sales")
data1.groupby("Segment")["Profit"].sum().sort_values(ascending=False)[:10].plot(kind="bar",color="lightgreen",ax=axes[1]).set_title('Segment Wise profit')
axes[1].set_ylabel("Profit")
```

```python
fig, axes = plt.subplots(1,2, figsize = (15, 5))
data1.groupby("City")["Sales"].sum().sort_values(ascending=False)[:10].plot(kind="bar",color="Teal",ax=axes[0]).set_title('Top 10 cities by Sales')
axes[0].set_ylabel("Sales")
data1.groupby("City")["Profit"].sum().sort_values(ascending=False)[:10].plot(kind="bar",color="tomato",ax=axes[1]).set_title('Top 10 cities by Profit')
axes[1].set_ylabel("Profit")
```

```python
fig, axes = plt.subplots(1,2, figsize = (15, 5))
data1.groupby("State")["Sales"].sum().sort_values(ascending=False)[:10].plot(kind="bar",color="crimson",ax=axes[0]).set_title('Top 10 State by Sales')
axes[0].set_ylabel("Sales")
data1.groupby("State")["Profit"].sum().sort_values(ascending=False)[:10].plot(kind="bar",color="indigo",ax=axes[1]).set_title('Top 10 State by Profit')
axes[1].set_ylabel("Profit")
```

```python
fig, axes = plt.subplots(1,2, figsize = (15, 5))
data1.groupby("Category")["Sales"].sum().plot(kind="pie",autopct='%1.1f%%',explode=[0,0,0.1],shadow=True,ax=axes[0]).set_title('Most Sales in Category')
data1.groupby("Category")["Profit"].sum().plot(kind="pie",autopct='%1.1f%%',explode=[0,0,0.1],shadow=True,ax=axes[1]).set_title('Most Profit in Category')
```

```python
fig, axes = plt.subplots(1,2, figsize = (15, 5))
data1.groupby("Sub-Category")["Sales"].sum().sort_values(ascending=False)[:10].plot(kind="bar",color="deepskyblue",ax=axes[0]).set_title('Most Sales in Sub-Category')
axes[0].set_ylabel("Sales")
data1.groupby("Sub-Category")["Profit"].sum().sort_values(ascending=False)[:10].plot(kind="bar",color="lime",ax=axes[1]).set_title('Most Proft in Sub-Category')
axes[1].set_ylabel("Profit")
```

```python
#scatter plot
plt.scatter(data1['Sales'],data1['Profit'], c ='green' )
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.title('Sales v/s Profit')
plt.show()

```

```python
fig, axes = plt.subplots(2,2, figsize = (16, 8))
fig.suptitle("Distribution Plots", fontsize = 16)
sns.distplot(data1['Sales'], ax=axes[0,0])
sns.distplot(data1['Profit'], ax=axes[0,1])
sns.distplot(data1['Discount'], ax=axes[1,0])
sns.distplot(data1['Quantity'], ax=axes[1,1])
```
